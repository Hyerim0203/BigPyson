```python
import pandas as pd
import numpy as np
from datetime import datetime,timedelta
from math import ceil
from dateutil.parser import parse
```

```python
df_rating = pd.read_excel('C:/Users/SM-PC/Desktop/01_제공데이터/2019시청률데이터.xlsx',header=1,index_col='시간대')
```

```python
# 양쪽 맨 끝에 있는 평균값 제거하기
df_rating = df_rating.iloc[:1440]
df_rating = df_rating.iloc[:,:-1]
```

```python

# 데이터 index와 column을 합친다
date = []
rating = []
for day in df_rating.columns:
    for time in df_rating.index:
        date.append('{} {}'.format(day,time))
        rating.append(df_rating.loc[time,day])
        
df_time = pd.DataFrame({'date':date,'rating':rating})
df_time
```

```python

# 빠른 실행을 위하여 0값 드랍
df_time = df_time[df_time.rating !=0]
df_time
```

```python

# 홈쇼핑 데이터와 시간 기준을 같게 만들고 str을 datetime으로 변경
def ys(date):
    if date[-5:]<'02:00':
        return datetime.strptime(date,'%Y-%m-%d %H:%M') + timedelta(days=1)
    else:
        return datetime.strptime(date,'%Y-%m-%d %H:%M')
df_time.date = df_time.date.apply(ys)
```

```python
df_time.sort_values('date')
```

```python
# date를 인덱스로 사용
df_time.set_index('date',inplace=True)
```

```python
#홈쇼핑실적데이터
df_ns = pd.read_excel('C:/Users/SM-PC/Desktop/01_제공데이터/2019실적데이터.xlsx')
                         
df_ns.rename(columns = {'방송일시':'date', '노출(분)':'exp_mins', '마더코드':'mom_code', '상품코드':'pd_code',
                 '상품명':'pd_name', '상품군':'pd_group', '판매단가':'pd_price', '취급액':'sales'}, inplace=True)
df_ns
```

```python
#무형 제거
df_excel_drop = df_excel.drop(index=df_excel[(df_excel['pd_group'] == '무형') & (df_excel['sales'].isnull())].index)
df_excel_drop
```

```python
#같은 시간대 같은 상품의 노출(분)을 사용하여 Null값을 채워주자.
#fillna의 메소드 중 ffill을 이용하여 Null을 채워주자.
#ffill 은 Null값의 바로 전 값을 사용해 Null을 채운다.
df_mom_code_ffill = df_excel_drop.fillna(method = 'ffill')
df_mom_code_ffill.info()
```

```python
# 편성표에 맞추어 시청률 평균값 구하기
#근데 이부분 잘 이해 안감,,,시간 올림이 무슨 소리일까
def rating_TV(row,df_time):
    st = str(row.date)
    fi = str(row.date + timedelta(minutes=ceil(row.exp_mins))) # 시간을 올림하였다.
    return( df_time[st:fi].sum())/ceil(row.exp_mins) # 0값을 날렸기에 더하고 기간으로 나눈다.
df_mom_code_ffill['rating'] = df_mom_code_ffill.apply(rating_TV,df_time=df_time,axis=1)
```

```python
#csv저장하기
df_mom_code_ffill.to_csv('C:/Users/SM-PC/Desktop/01_제공데이터/가공데이터/실적+시청률.csv', index = False, encoding='utf-8-sig')
```
